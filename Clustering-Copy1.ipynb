### 9.Build dummy variables for categorical variables

### 10.Get x as columns of ApplicantIncome and Credit_History


```python
X=test.iloc[:,[0,4]]
```

```python
X
```

### 11.Import matplotlib


```python
import matplotlib.pyplot as plt
```

### 12.Elbow method to determine K


```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### 13.fit K_means with founded K



```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

### 14.show cluster results


```python
y_kmeans
```

### 15.print cluster centers


```python
kmeans.cluster_centers_
```

### 16.Lets plot those clusters!


```python
fig = plt.figure(figsize = (10,5))
plt.scatter(X[y_kmeans==0], X[y_kmeans==0], s=100, c='yellow',   label='group1')
plt.scatter(X[y_kmeans==1], X[y_kmeans==1], s=100, c='red',     label='group2')
plt.scatter(X[y_kmeans==2], X[y_kmeans==2], s=100, c = 'blue', label = 'group3')
plt.scatter(X[y_kmeans==3], X[y_kmeans==3], s=100, c = 'black', label = 'group4')
plt.title('Clustering')
plt.xlabel('ApplicantIncome')
plt.ylabel('Credit_History')
plt.legend()
plt.show()
```

### 18.Fit agglomerative hierarchical clustering



```python
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering


ac = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
hc_predict = ac.fit_predict(X_)
```
